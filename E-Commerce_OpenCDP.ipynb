```python
def load_ecommerce_data(month):
    path = f"/Volumes/workspace/ecommerce/ecommerce_data/2019-{month}.csv"
    
    df = spark.read.option("header", "true").option("inferSchema", "true") \
        .csv(path)
    
    return df
```

```python
events = load_ecommerce_data("Oct")
display(events.limit(5))
```

```python
# Create simple DataFrame
data_sample = [("iPhone", 999), ("Samsung", 799), ("MacBook", 1299)]
df1 = spark.createDataFrame(data_sample, ["product", "price"])
df1.show()

# Filter expensive products
df1.filter(df1.price > 1000).show()
```
