**Read data into Dataframe**

```python
# Load data
events_oct = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv", header=True, inferSchema=True)
```

```python
display(events_oct.limit(5))
```

```python
events_oct.printSchema
```

```python
# Basic operations
events_oct.select("event_type", "brand", "price").show(10)
```

```python
events_oct.filter("price > 100").count()
```

```python
events_oct.groupBy("event_type").count().show()
```

```python
top_brands = events_oct.groupBy("brand").count().orderBy("count", ascending=False).limit(5)
top_brands.show()
```
