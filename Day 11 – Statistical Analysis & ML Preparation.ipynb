### 🎯 Goal - Understand distribution, spread, and central tendency 

```python
%sql
SELECT
  COUNT(*) AS total_events,
  AVG(price) AS avg_price,
  MIN(price) AS min_price,
  MAX(price) AS max_price,
  STDDEV(price) AS stddev_price
FROM ecommerce.silver.events
WHERE price IS NOT NULL;
```

### Price stats by category

```python
%sql
SELECT
  category_code,
  COUNT(*) AS cnt,
  ROUND(AVG(price), 2) AS avg_price,
  ROUND(STDDEV(price), 2) AS price_std
FROM ecommerce.silver.events
WHERE price IS NOT NULL
GROUP BY category_code
ORDER BY cnt DESC;
```

```python
%sql
WITH labeled AS (
  SELECT
    price,
    CASE
      WHEN dayofweek(event_time) IN (1,7) THEN 'Weekend'
      ELSE 'Weekday'
    END AS day_type
  FROM ecommerce.silver.events
  WHERE event_type = 'purchase'
)
SELECT
  day_type,
  COUNT(*) AS purchases,
  ROUND(AVG(price), 2) AS avg_spend
FROM labeled
GROUP BY day_type;
```

### Correlation: user activity vs spend

```python
%sql
WITH user_metrics AS (
  SELECT
    user_id,
    COUNT(*) AS total_events,
    SUM(CASE WHEN event_type = 'purchase' THEN price ELSE 0 END) AS total_spent
  FROM ecommerce.silver.events
  GROUP BY user_id
)
SELECT
  CORR(total_events, total_spent) AS corr_activity_spend
FROM user_metrics;
```

```python
%sql
CREATE OR REPLACE TABLE ecommerce.gold.user_features AS
SELECT
  user_id,COUNT(*) AS total_events,
  SUM(CASE WHEN event_type = 'view' THEN 1 ELSE 0 END) AS views,
  SUM(CASE WHEN event_type = 'cart' THEN 1 ELSE 0 END) AS carts,
  SUM(CASE WHEN event_type = 'purchase' THEN 1 ELSE 0 END) AS purchases,
  ROUND(SUM(CASE WHEN event_type = 'purchase' THEN price ELSE 0 END), 2) AS total_spent,
  COUNT(DISTINCT user_session) AS sessions,
  MAX(event_time) AS last_activity_ts
FROM ecommerce.silver.events
GROUP BY user_id;
```

```python
%sql
CREATE OR REPLACE TABLE ecommerce.gold.user_features_ml AS
SELECT
  *,
  purchases * 1.0 / NULLIF(views, 0)        AS view_to_purchase_rate,
  purchases * 1.0 / NULLIF(sessions, 0)     AS purchase_per_session
FROM ecommerce.gold.user_features;
```

```python
%sql
ALTER TABLE ecommerce.gold.user_features_ml
ADD COLUMNS (days_since_last_activity INT);

UPDATE ecommerce.gold.user_features_ml
SET days_since_last_activity =
  DATEDIFF(CURRENT_DATE(), DATE(last_activity_ts));
```

```python
from pyspark.sql import functions as F

events_wd = events.withColumn(
    "is_weekend",
    F.dayofweek("event_time").isin([1, 7])
)

events_wd.groupBy("is_weekend").agg(
    F.sum(F.when(F.col("event_type") == "view", 1).otherwise(0)).alias("views"),
    F.sum(F.when(F.col("event_type") == "purchase", 1).otherwise(0)).alias("purchases"),
    (F.sum(F.when(F.col("event_type") == "purchase", 1).otherwise(0)) /
     F.sum(F.when(F.col("event_type") == "view", 1).otherwise(0))
    ).alias("conversion_rate")
).show()
```

```python
user_metrics = events.groupBy("user_id").agg(
    F.count("*").alias("total_events"),
    F.sum(F.when(F.col("event_type") == "purchase", F.col("price")).otherwise(0)).alias("total_spent")
)

user_metrics.stat.corr("total_events", "total_spent")
```

```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

user_window = Window.partitionBy("user_id").orderBy("event_time")

features = events \
    .withColumn("hour", F.hour("event_time")) \
    .withColumn("day_of_week", F.dayofweek("event_time")) \
    .withColumn("price_log", F.log(F.col("price") + 1)) \
    .withColumn(
        "time_since_first_view",
        F.unix_timestamp("event_time") -
        F.unix_timestamp(F.first("event_time").over(user_window))
    )
```
